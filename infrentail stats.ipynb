```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets
bookseller_data = pd.read_csv('/content/booksellar.csv')
debugging_data = pd.read_csv('/content/debugging.csv')
```

```python
# 1. Bookseller Analysis
books_sold = bookseller_data['Number of Books Sold']
```

```python
# Probability for "more than or equal to 96 books"
prob_96_or_more = (books_sold >= 96).mean()
```

```python
# Probability for "less than or equal to 92 books"
prob_92_or_less = (books_sold <= 92).mean()
```

```python
# Visualization: Distribution of books sold
plt.figure(figsize=(10, 6))
sns.histplot(books_sold, kde=True, bins=15, color='skyblue', alpha=0.7)
plt.axvline(x=96, color='green', linestyle='--', label='96 Books (Boundary)')
plt.axvline(x=92, color='red', linestyle='--', label='92 Books (Boundary)')
plt.title("Distribution of Books Sold per Day")
plt.xlabel("Number of Books Sold")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# 2. Debugging Analysis
time_taken = debugging_data['Time Taken to fix the bug']
```

```python
# Probability for "less than 3 hours"
prob_less_than_3 = (time_taken < 3).mean()
```

```python
# Probability for "more than 2 hours"
prob_more_than_2 = (time_taken > 2).mean()
```

```python
# 50th Percentile (Median)
median_time = np.percentile(time_taken, 50)
```

```python
# Visualization: Distribution of time taken to fix bugs
plt.figure(figsize=(10, 6))
sns.histplot(time_taken, kde=True, bins=15, color='lightcoral', alpha=0.7)
plt.axvline(x=3, color='blue', linestyle='--', label='3 Hours (Boundary)')
plt.axvline(x=2, color='green', linestyle='--', label='2 Hours (Boundary)')
plt.axvline(x=median_time, color='purple', linestyle='--', label=f'Median: {median_time:.2f} Hours')
plt.title("Distribution of Time Taken to Fix Bugs")
plt.xlabel("Time Taken to Fix Bugs (Hours)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Print Results
print("Bookseller Analysis:")
print(f"Probability (>= 96 books): {prob_96_or_more:.3f}")
print(f"Probability (<= 92 books): {prob_92_or_less:.3f}")
```

```python
print("\nDebugging Analysis:")
print(f"Probability (< 3 hours): {prob_less_than_3:.3f}")
print(f"Probability (> 2 hours): {prob_more_than_2:.3f}")
print(f"50th Percentile (Median): {median_time:.3f} hours")
```
